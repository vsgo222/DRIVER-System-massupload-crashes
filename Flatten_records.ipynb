```python
def flatten_json(y):
    out = {}
    
    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[str(name[:-1])] = str(x)
        
    flatten(y)
    return out

# for ind in data_df.index:  
flat = flatten_json(data_df['data'].values.tolist())
flat

# # df_data = pd.DataFrame.from_dict(flat)
# # df_data
# # df_data.to_csv('flattened.csv')
```
